```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 50)
```

```python
df = pd.read_csv('hotel_booking.csv')
```

```python
df
```

```python
df.columns
```

```python
df = df.drop(columns=[
    'arrival_date_week_number', 'market_segment', 'booking_changes',
    'company', 'days_in_waiting_list', 'total_of_special_requests',
    'email', 'phone-number', 'credit_card'
])
```

```python
month_string_to_num_map = dict(zip(pd.date_range('2000-01-01', freq='M', periods=12).strftime('%B'), range(1,13)))
month_string_to_num_map
```

```python
df['arrival_date_month'] = df['arrival_date_month'].map(month_string_to_num_map)
```

```python
df
```

```python
df['arrival_date_year'] = pd.to_datetime(
    df[['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month']]
        .astype(str)
        .agg('/'.join, axis=1)
)

df.drop(columns=['arrival_date_month', 'arrival_date_day_of_month'], inplace=True)
df.rename(columns={'arrival_date_year': 'arrival_date'}, inplace=True)
```

```python
df['children'] = df['children'].astype(pd.Int64Dtype())
df['agent'] = df['agent'].astype(pd.Int64Dtype())
```

```python
df
```

```python
df.to_csv('hotel_booking_cleaned.csv', index=None)
```
